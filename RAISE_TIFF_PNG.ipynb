```python
import os
import requests
import pandas as pd
from PIL import Image
from io import BytesIO
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load CSV
csv_path = "dataset/RAISE-1k/RAISE_1k.csv"   # <-- update path if needed
df = pd.read_csv(csv_path)

# Set base_dir inside Google Drive
base_dir = "/content/drive/MyDrive/RAISE"   # saves in your Drive
tiff_dir = os.path.join(base_dir, "TIFF")
png_dir = os.path.join(base_dir, "PNG")

os.makedirs(tiff_dir, exist_ok=True)
os.makedirs(png_dir, exist_ok=True)

# Function to download TIFF and convert to PNG
def download_and_convert(file_id, tiff_url):
    tiff_path = os.path.join(tiff_dir, f"{file_id}.TIF")
    png_path = os.path.join(png_dir, f"{file_id}.png")

    try:
        # Download TIFF
        response = requests.get(tiff_url, timeout=30)
        response.raise_for_status()
        with open(tiff_path, "wb") as f:
            f.write(response.content)
        print(f"‚úÖ Saved TIFF: {tiff_path}")

        # Convert TIFF -> PNG
        image = Image.open(BytesIO(response.content))
        image.save(png_path, "PNG")
        print(f"üñºÔ∏è Converted to PNG: {png_path}")

    except Exception as e:
        print(f"‚ùå Failed for {file_id}: {e}")

# ======= TEST MODE: process only first N images =======
N = 5   # change this number to test with more/less images
for idx, row in df.head(N).iterrows():
    file_id = row["File"]
    tiff_url = row["TIFF"]
    download_and_convert(file_id, tiff_url)
```

```python
from google.colab import drive
import os
import requests
import pandas as pd
from PIL import Image
from io import BytesIO
import time

# ====== Mount Google Drive ======
drive.mount('/content/drive')

# Base directory in Drive (only PNGs will be saved here)
base_dir = "/content/drive/MyDrive/RAISE"
png_dir = os.path.join(base_dir, "PNG")
os.makedirs(png_dir, exist_ok=True)

# Load CSV (upload RAISE_1k.csv or put it in Drive)
csv_path = "dataset/RAISE-1k/RAISE_1k.csv"
df = pd.read_csv(csv_path)

# Function: download TIFF, convert to PNG, save PNG only
def download_and_convert(file_id, tiff_url):
    png_path = os.path.join(png_dir, f"{file_id}.png")

    # Skip if PNG already exists
    if os.path.exists(png_path):
        print(f"‚è© Skipped (already exists): {png_path}")
        return

    try:
        # Download TIFF into memory
        response = requests.get(tiff_url, timeout=60)
        response.raise_for_status()

        # Convert TIFF -> PNG
        image = Image.open(BytesIO(response.content))
        image.save(png_path, "PNG")
        print(f"‚úÖ Saved PNG: {png_path}")

    except Exception as e:
        print(f"‚ùå Failed for {file_id}: {e}")

# ====== Batch processing ======
batch_size = 100
total_images = len(df)

for start in range(0, total_images, batch_size):
    end = min(start + batch_size, total_images)
    print(f"\nüöÄ Processing batch {start+1} ‚Üí {end} of {total_images}")

    for idx, row in df.iloc[start:end].iterrows():
        file_id = row["File"]
        tiff_url = row["TIFF"]
        download_and_convert(file_id, tiff_url)

    print(f"‚úÖ Finished batch {start+1} ‚Üí {end}")
    # Optional small delay to reduce server load
    time.sleep(5)
```
